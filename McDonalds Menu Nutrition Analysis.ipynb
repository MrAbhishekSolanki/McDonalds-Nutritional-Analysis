# Objectives

The primary objectives of this Exploratory Data Analysis (EDA) project on the McDonald's dataset are as follows:

**Nutrient Distribution by Category:**
Examine and visualize the distribution of key nutrients, including carbohydrates, proteins, and fats, across different food categories within the McDonald's menu.

Investigate and plot the total sugar content in various food categories to understand the variation in sugar levels across different menu items.

Explore and represent the total sodium content within each food category to identify categories with varying sodium levels.

These objectives aim to provide a comprehensive overview of the nutritional composition, diversity, and characteristics of menu items within each category of the McDonald's dataset. The analysis contributes to a better understanding of the overall nutritional profile and composition of McDonald's offerings.

#### What are the number of Items in each category?

```python
#sns.set(style="whitegrid")
ax = sns.countplot(df['Category'])

for bars in ax.containers:
    ax.bar_label(bars)
```

Observation : Coffee & Tea has most variety of items in total and Salads and Desserts have least variety of items in total. 

#### What are the values of average calories in each category of McDonald's menu?

```python
average_calories_per_cat = df.groupby('Category')['Calories'].mean().reset_index()
average_calories_per_cat = average_calories_per_cat.sort_values('Calories', ascending=False)
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(x = 'Category',y ='Calories', data=average_calories_per_cat, palette='viridis', hue='Category')
plt.xticks(rotation=85)
plt.show()
```

Observation: Chicken & Fish Category has the highest observation in Calories and Beverages has the lowest. 

#### What is the nutrient distribution(Carbs, proteins, fats) in each category?

```python
df.columns
```

```python
nutrients_dist = df.groupby('Category')[['Carbohydrates','Protein','Total Fat']].mean().round().sort_values(by=['Carbohydrates','Protein','Total Fat'], ascending=[False,False,False])
```

```python
nutrients_dist.plot(kind='bar', color=['blue','green','yellow'] )
plt.title("Nutrients distribution in each category")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

So from above graph, we can oberve that the top categories that share the most amount of Nutrients are: 
1. Carbs -> Smoothies & Shakes
2. Proteins -> Chicken and Fish
3. Total Fats -> Breakfast

While, the categories with the least amount of nutrients are: 
1. Carbs -> Salads
2. Proteins -> Beverages
3. Total Fats -> Beverages

#### What is the total sugar content measured by Categories?

```python
Total_sug_by_cat = pd.DataFrame(df.groupby('Category')['Sugars'].sum().sort_values(ascending=False))
Total_sug_by_cat.plot(kind='bar')
plt.ylabel("Total Sugar Count")
plt.title("Total Sugars visualized by categories")
plt.show()
```

From this graph, we can see that the Category "Coffee & Tea" has the most amount of added, followed by Smoothies and Shakes, and Salads contain the least amount of sugar. 

#### What is the total Sodium content measured by Categories?

```python
total_sodium_per_cat = pd.DataFrame(df.groupby('Category')['Sodium'].sum().sort_values(ascending=False))
total_sodium_per_cat
```

```python
total_sodium_per_cat.plot(kind='bar', color='grey')
plt.title("Total Sodium present per category")
plt.ylabel('Values')
plt.xticks(rotation=90)
plt.show()
```

Now from this last graph, we can observe that breakfast contains the most amount of sodium, followed by chicken and fish. While, desserts contain the least amount of sodium amongst all. 

#### What are the average cholestrol values per food category?

```python
df.columns
```

```python
cholestrol_values = df.groupby('Category')['Cholesterol'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))
cholestrol_values.plot(kind='bar')
plt.title('Cholestrol Values visualized per category')
plt.ylabel('Avg Cholestrol(mg)')
plt.show
```

From the above graph we can conclude that, the Category 'Breakfast' alone far exceeds in cholestrol count compared to other food categories, ie: exceeds the second most cholestrol containing category(Beef and Pork) by 72%. While desserts and beverages contain the least amount of cholestrol.
